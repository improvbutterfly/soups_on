```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
soup = pd.read_csv("Resources/soup_data.csv")
soup.head()
```

```python
popular_voters_count = soup.sort_values(["voters","rating"], ascending=False)
popular_voters_count
```

```python
popularity_by_rating = soup.sort_values(["rating","voters"], ascending=False)
popularity_by_rating.loc[popularity_by_rating["voters"]>10,:]
```

```python
# Scatter Plot rating and voters
popularity_by_rating.loc[(popularity_by_rating["rating"]>0) & (popularity_by_rating["voters"]>10),["rating","voters"]]\
                    .plot("rating","voters",kind="scatter", figsize=(6,4),
                         title="Soup recipes ratings vs. number of voters")
plt.show()
plt.tight_layout()
```

```python
# Scatter Plot rating and number of ingredients
popularity_by_rating.loc[(popularity_by_rating["rating"]>0) & (popularity_by_rating["voters"]>10),["rating","n_ingredients"]]\
                    .plot("rating","n_ingredients",kind="scatter", figsize=(6,4),
                         title="Soup recipes ratings vs. number of ingredients")
plt.show()
plt.tight_layout()
```

```python
# Scatter Plot rating and minutes to cook
popularity_by_rating.loc[(popularity_by_rating["rating"]>0) & (popularity_by_rating["voters"]>10) \
                         & (popularity_by_rating["minutes"]<1000),["rating","minutes"]]\
                    .plot("rating","minutes",kind="scatter", figsize=(6,4),
                         title="Soup recipes ratings vs. minutes to cook")
plt.show()
plt.tight_layout()
```

```python
# Scatter Plot number of ingredients and minutes to cook
popularity_by_rating.loc[(popularity_by_rating["rating"]>0) & (popularity_by_rating["voters"]>10)\
                         & (popularity_by_rating["minutes"]<1000),["minutes","n_ingredients"]]\
                    .plot("n_ingredients","minutes",kind="scatter", figsize=(6,4),
                         title="Soup recipes number of ingredients vs. minutes to cook")
plt.show()
plt.tight_layout()
```
